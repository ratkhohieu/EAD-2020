```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os.path as osp
import sys
import cv2
import matplotlib.pyplot as plt


EAD_CLASSES = (  # always index 0
    'specularity','saturation','artifact',
    'blur','contrast','bubbles','instrument','blood')

def __class_to_ind__(class_name):
    class_to_ind = dict(zip(EAD_CLASSES, range(len(EAD_CLASSES))))
    ind = class_to_ind[class_name]
    return ind


def __txt2list__(path):
    res = []
    for line in open(path):
        name_laber = line.split(' ')[0]
        label_ind = __class_to_ind__(name_laber)

        xmin = int(line.split(' ')[1])
        ymin = int(line.split(' ')[2])
        xmax = int(line.split(' ')[3])
        ymax = int(line.split(' ')[4])

        bndbox = [xmin, ymin, xmax, ymax, label_ind]
        res += [bndbox]
    return res  # [[xmin, ymin, xmax, ymax, label_ind], ... ]


def load_image(path):
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image


class EADDetection(Dataset):

    def __init__(self, root, transform=None):
        self.root = root
        self.transform = transform
        self._txtpath = osp.join('./data/EAD2020_dataType_framesOnly/yolo2voc', '%s.txt')
        self._imgpath = osp.join('./data/EAD2020_dataType_framesOnly/frames', '%s.jpg')
        self.ids = list()
        for line in open(self.root):
            self.ids.append(line.strip())

    def __getitem__(self, index):
        img_id = self.ids[index]

        img = load_image(self._imgpath % img_id)
        height, width, channels = img.shape

        target = __txt2list__(self._txtpath % img_id)
        target = np.array(target)
        bbox = target[:, :4]
        labels = target[:, 4]

        if self.transform is not None:
            annotation = {'image': img, 'bboxes': bbox, 'category_id': labels}
            augmentation = self.transform(**annotation)
            img = augmentation['image']
            bbox = augmentation['bboxes']
            labels = augmentation['category_id']
        return {'image': img, 'bboxes': bbox, 'category_id': labels}

    def __len__(self):
        return len(self.ids)

    def __num_class__(self):
        return len(EAD_CLASSES)

    def label_to_name(self, label):
        return EAD_CLASSES[label]
```

```python
import torch
from torch.utils.data import Dataset, DataLoader
import numpy as np
import os.path as osp
import sys
import cv2
import matplotlib.pyplot as plt


EAD_CLASSES = (  # always index 0
    'specularity','saturation','artifact',
    'blur','contrast','bubbles','instrument','blood')

def __class_to_ind__(class_name):
    class_to_ind = dict(zip(EAD_CLASSES, range(len(EAD_CLASSES))))
    ind = class_to_ind[class_name]
    return ind


def __txt2list__(path):
    res = []
    for line in open(path):
        name_laber = line.split(' ')[0]
        label_ind = __class_to_ind__(name_laber)

        xmin = int(line.split(' ')[1])
        ymin = int(line.split(' ')[2])
        xmax = int(line.split(' ')[3])
        ymax = int(line.split(' ')[4])

        bndbox = [xmin, ymin, xmax, ymax, label_ind]
        res += [bndbox]
    return res  # [[xmin, ymin, xmax, ymax, label_ind], ... ]


def load_image(path):
    image = cv2.imread(path)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
    return image


class EADDetection(Dataset):

    def __init__(self, root, transform=None, dataset_name='EAD20'):
        self.root = root
        self.transform = transform
        self.name = dataset_name
        self._txtpath = osp.join('./data/EAD2020_dataType_framesOnly/yolo2voc', '%s.txt')
        self._imgpath = osp.join('./data/EAD2020_dataType_framesOnly/frames', '%s.jpg')
        self.ids = list()
        for line in open(self.root):
            self.ids.append(line.strip())



    def __getitem__(self, index):
        im, gt, h, w = self.pull_item(index)

        return im, gt


    def pull_item(self, index):
        img_id = self.ids[index]
        img = cv2.imread(self._imgpath % img_id)
        #img = load_image(self._imgpath % img_id)
        height, width, channels = img.shape
        print(img)
        target = __txt2list__(self._txtpath % img_id)

        if self.transform is not None:
            target = np.array(target)
            img, boxes, labels = self.transform(img, target[:, :4], target[:, 4])
            # to rgb
            img = img[:, :, (2, 1, 0)]
            # img = img.transpose(2, 0, 1)
            target = np.hstack((boxes, np.expand_dims(labels, axis=1)))
        return torch.from_numpy(img).permute(2, 0, 1), target, height, width
        # return torch.from_numpy(img), target, height, width

    # def __getitem__(self, index):
    #     img_id = self.ids[index]
    #
    #     img = load_image(self._imgpath % img_id)
    #     height, width, channels = img.shape
    #
    #     target = __txt2list__(self._txtpath % img_id)
    #     target = np.array(target)
    #     bbox = target[:, :4]
    #     labels = target[:, 4]
    #
    #     if self.transform is not None:
    #         annotation = {'image': img, 'bboxes': bbox, 'category_id': labels}
    #         augmentation = self.transform(**annotation)
    #         img = augmentation['image']
    #         img = np.transpose(img, (2, 0, 1)).astype(np.float32)
    #         bbox = augmentation['bboxes']
    #         labels = augmentation['category_id']
    #     return {'image': img, 'bboxes': bbox, 'category_id': labels}

    def __len__(self):
        return len(self.ids)

    def __num_class__(self):
        return len(EAD_CLASSES)

    def label_to_name(self, label):
        return EAD_CLASSES[label]
```
